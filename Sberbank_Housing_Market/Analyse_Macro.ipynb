# macro.csv

A set of macroeconomic indicators, one for each date.

- eurrub: Ruble/EUR exchange rate

- gdp_annual: GDP at current prices

- deposits_rate: Average interest rate on deposits

- mortgage_growth: Growth of mortgage lending
- mortgage_rate: Weighted average rate of mortgage loans

- grp_growth: Growth of gross regional product of the subject of the Russian Federation where Apartment is located

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
import re
import scipy
from collections import defaultdict
```

```python
def print_correlation_matrix(df):
    #print correlation matrix
    f, ax = plt.subplots(figsize=(20, 20))
    corr = df.corr()
    sns.heatmap(corr, square=True, ax=ax)
    
def print_top_correlation_indices(df): 
    #print highest correlated variables
    corr = df.corr()
    corr.loc[:,:] = np.tril(corr.values, k=-1) #lower triangular
    sorted_corr = corr.abs().unstack().sort_values(ascending = False)

    pd.set_option("display.max_rows",500)
    print(sorted_corr[sorted_corr > 0.8])
    
def get_columns_satisfying_regex(regex_str):
    this_regex = re.compile(regex_str)
    return [m.group(0) for l in column_names for m in [this_regex.search(l)] if m]

def convert_string_to_datetime(string):
    return pd.to_datetime(string, format='%Y-%m-%d', errors='ignore')
```

# Imputing missing values
First thing we note is that macro only changes every month. Even though we have daily data, it only changes on the first of every month.
Now we check which columns have NAN.

```python
macro_df = macro_df.drop(macro_columns_to_delete, 1)
test_macro = test_macro.drop(macro_columns_to_delete, 1)
num_test_days = test_macro.shape[0]
filled_days = test_macro.count()
```

```python
macro_df.columns.tolist()
```

```python
columns_to_input = filled_days[filled_days < num_test_days].index.tolist()
```

```python
columns_to_input
```

```python
main_macro = macro_df[macro_df['timestamp'].apply(lambda x: x.day == 1)]
year_macro = main_macro[main_macro['timestamp'].apply(lambda x: x.month == 1)]
```

Check who changes monthly and who changes yearly

```python
macro_cols = macro_df.drop('timestamp',1).columns.values.tolist()
month_cols = []
year_cols = []

for var in macro_cols:
    num_values = main_macro.groupby(var)["timestamp"].unique().shape[0]
    if num_values < 7:
        year_cols += [var]
    else:
        month_cols += [var]
```

```python
year_cols
```

```python
month_cols
```

All years except 2016 have all its information filled. The following columns have na:

```python
null_2016 = main_macro[main_macro['timestamp'].apply(lambda x:(x.year == 2016))]
null_2016 = null_2016.columns[null_2016.isnull().any()].tolist()
null_2016
```

 - Variables that change monthly will get the average value of the previous year
 - Variables that change year will get the average increase from all years
 
## Monthly changes

```python
def get_final_known_value(column):
    only_non_na = main_macro[main_macro[column].notnull()]
    final_row = only_non_na.tail(1)[column]
    return final_row.values.tolist()[0]
```

```python
def compute_tick(time):
    if time.month == 1:
        return str(time.year)
    else:
        return ""

tick_labels = main_macro['timestamp'].apply(lambda x: compute_tick(x), 0)
```

```python
set.intersection(set(month_cols), set(null_2016))
```

### income_per_cap

```python
g = sns.pointplot(data = main_macro, x = 'timestamp', y = 'income_per_cap', ci = 0)
g.set(xticklabels=tick_labels); #; hides prints
```

Here maybe we impute by the mean of the previous year

```python
mean_cap_2015 = main_macro[main_macro['timestamp'].apply(lambda x: x.year == 2015)]['income_per_cap'].mean()
macro_df['income_per_cap'] = macro_df['income_per_cap'].fillna(mean_cap_2015)
```

### deposits_rate

```python
g = sns.pointplot(data = main_macro, x = 'timestamp', y = 'deposits_rate', ci = 0)
g.set(xticklabels=tick_labels); #; hides prints
```

2010 and some final 2016 days doesn't have values. So we impute these dates with the closest date with the value known. So 2010 will have the same value as the first day of 2011, and 2016 the same value as late-2016.

```python
first_deposits_rate_value = main_macro[main_macro['deposits_rate'].notnull()].head(1)['deposits_rate'].values.tolist()[0]

macro_2010 = macro_df[macro_df['timestamp'].apply(lambda x: x.year == 2010)]
macro_df.loc[macro_2010.index,'deposits_rate'] = macro_2010['deposits_rate'].fillna(first_deposits_rate_value)


last_deposits_rate_value = get_final_known_value('deposits_rate')

macro_2016 = macro_df[macro_df['timestamp'].apply(lambda x: x.year == 2016)]
macro_df.loc[macro_2016.index,'deposits_rate'] = macro_2016['deposits_rate'].fillna(last_deposits_rate_value)
```

## yearly changes

```python
tick_labels_year = year_macro['timestamp'].apply(lambda x: x.year, 0)
```

```python
year_null_2016 = set.intersection(set(year_cols), set(null_2016))
```

```python
g = sns.pointplot(data = year_macro, x = 'timestamp', y = 'salary', ci = 0)
g.set(xticklabels=tick_labels_year); #; hides prints
```

```python
year_null_2016
```

Here we are going to add by the average of increases.

```python
def get_predicted_value(column):
    average_increase = year_macro[column].diff().mean()
    return get_final_known_value(column) + average_increase

for var in year_null_2016:
    value = get_predicted_value(var)
    macro_df[var] = macro_df[var].fillna(value)
```
